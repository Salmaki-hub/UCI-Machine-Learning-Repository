# Now removing the irrelevant columns in the dataset and performing EDA analysis

```python
# Drop irrelevant columns
# I'm dropping the columns that I find have minimum or no impact on the Target.
# The columns I'll keep are "All ciricular columns", "Martial Status", "Course", "Previous Qualification","Gender" & "Target"
Data = Data.drop(['Application mode',	'Application order','Daytime/evening attendance\t','Previous qualification (grade)','Nacionality',"Mother's qualification","Father's qualification","Mother's occupation","Father's occupation",'Admission grade','Displaced','Educational special needs','Debtor','Tuition fees up to date','Scholarship holder','Age at enrollment','International'], axis=1)
Data.head(5)
```

```python
# Total number of rows and columns
Data.shape
```

```python
# Count the rows before removing the data
Data.count()
```

```python
# Rows containing duplicate data
duplicate_rows = Data[Data.duplicated()]
print("Number of duplicate rows: ", duplicate_rows.shape)
```

```python
# Drop the duplicates
Dataset = Data.drop_duplicates()
Dataset.head()
```

```python
## Count the number of rows after removing duplicates.
Dataset.count()
```

```python
# Find the null values
print(Dataset.isnull().sum())
```

```python
# Drop the missing values.
Dataset = Dataset.dropna()
Dataset.count()
```

```python
# Finding the outliers
sns.scatterplot(x="Target",y="Course",data = Dataset)
```

```python
sns.boxplot(x=Dataset['Previous qualification'])
```

```python
# Plotting a Histogram for number of cars per brand
Dataset.Target.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of students by target")
plt.ylabel("Number of students")
plt.xlabel("Target");
```

```python
# Find the relations between the variables.
plt.figure(figsize=(20,10))
c= Dataset.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
#Relationship analysis
sns.relplot(x = "Curricular units 1st sem (credited)",y = "Curricular units 2nd sem (credited)", hue = 'Target', data = Dataset)
```
